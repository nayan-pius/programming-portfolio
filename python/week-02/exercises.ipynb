### Concept 1 — Lists


```python
networks = ["MTN", "Vodafone", "AirtelTigo"]
print(networks)
```

Now here is something important — you can access one item from a list using its index. Python starts counting from 0, not 1:

```python
print(networks[0])  # MTN
print(networks[1])  # Vodafone
print(networks[2])  # AirtelTigo
```

### Concept 2 — for loops
A for loop lets you go through every item in a list automatically — one by one — without writing the same line over and over.

#### Without a loop:


```python
networks = ["MTN", "Vodafone", "AirtelTigo"]
print(networks[0])
print(networks[1])
print(networks[2])
```

### With a loop
Same results, but shorter codes

```python
networks = ["MTN", "Vodafone", "AirtelTigo"]
for network in networks:
    print(network)
    
```

🌐In real networking you use loops to ping every device in a list:


```python
ip_addresses = ["192.168.1.1", "192.168.1.2", "192.168.1.3"]
for ip in ip_addresses:
    print(f"Pinging {ip}...")
```

### Concept 3 — while loops
A while loop keeps running as long as a condition is True. It does not loop through a list — it loops until something changes.

```python
count = 1

while count <= 5:
    print(f"Count is {count}")
    count = count + 1

# "While count is less than or equal to 5 — print it and add 1"
```

⚠️ Very important — always change the variable inside the loop. If you forget count = count + 1 the condition never becomes False and your program runs forever. This is called an infinite loop and it will freeze your terminal.


🌐In real networking a while loop can keep retrying a connection:

```python
attempts = 1

while attempts <= 3:
    print(f"Attempting to connect to MTN-Router... attempt {attempts}")
    attempts = attempts + 1

print("No more attempts.")
```
