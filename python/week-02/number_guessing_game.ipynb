# 📌 Project Overview
A command-line number guessing game where Python picks a random secret number between 1 and 10 and the user keeps guessing until they get it right. The game tracks how many attempts it took.

```python
import random
secret_number = random.randint(1,10)
attempts = 0
user_guess = int(input("Guess the Secret Number: "))

while secret_number != user_guess:
    attempts = attempts + 1
    if secret_number > user_guess:
        print("The guessed number is too low")
    elif secret_number < user_guess:
        print("The guessed number is too high")
    user_guess = int(input("Guess the Secret Number: "))

    
attempts = attempts + 1       
print("You guessed it correctly!🎉")
print(f"You tried {attempts} times")
```
