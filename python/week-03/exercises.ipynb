## Function
A function is a block of code you write once and can use as many times as you want. Instead of repeating the same code over and over, you put it in a function and just call it by name.

### Without a function:

```python
print("Welcome to MTN")
print("Welcome to MTN")
print("Welcome to MTN")
```

### With a function:

```python
def greet():
    print("Welcome to MTN")

greet()
greet()
greet()
```

Same output — but now the code lives in one place. If you want to change the message you change it once, not three times.

## Parameters and Arguments

Right now your functions do the same thing every time they are called. Parameters let you pass information into a function so it can behave differently each time.

```python
def greet(name):
    print(f"Welcome to MTN, {name}")

greet("Pius")
greet("Kofi")
greet("Ama")


# Same function — different result each time based on what you pass in.

# The vocabulary:
# name inside def greet(name): → this is the parameter — a placeholder
# "Pius" inside greet("Pius") → this is the argument — the actual value
```

### Multiple parameter

```python
def describe_device(device, ip):
    print(f"Device: {device} | IP: {ip}")

describe_device("MTN-Router-01","192.168.1.1")
describe_device("Vodafone-Switch","192.168.1.2")
```

```python
def ping_device(ip):
    print(f"Pinging {ip}...")

ping_device("192.168.1.1")
ping_device("192.168.1.2")
ping_device("192.168.1.3")
```

## Return Values

So far your functions just print things. But what if you want to use the result of a function later in your code? That is where return comes in.

### Without return:

```python
def add(a,b):
    print(a + b) # just prints, result is lost

add(5,3) # prints 8 but can't use it
```

### With return

```python
def add(a,b):
    return a+b # sends the result back

result = add(5,3) # result now holds 8
print(f"The answer is {result}")
```

```python
# 🌐In real networking:

def get_device_info(device, ip):
    return f"Device: {device} | IP: {ip}"

info = get_device_info("MTN-Router-01", "192.168.1.1")
print(info)
```

```python
def calculate_subnet(network,prefix):
    return f"Network: {network} | Prefix: {prefix}"

subnet = calculate_subnet("192.168.1.0","/24")
print(subnet)
```
