## Password Strength Checker
This program checks how strong a password is based on 4 rules:
1. Length — at least 8 characters
2. Contains a number
3. Contains a special character
4. Contains an uppercase letter<br>

                        Built by Pius | Week 3 Saturday | April 2026

### Rule 1
Checks the length of the password

```python
def check_length(password):
    if len(password) >= 8:
        return True
    else:
        return False
    
```

### Rule 2
Checks if the password includes number(s)

```python
def check_number(password):
    for number in password:
        if number.isdigit():
            return True
    return False    
```

### Rule 3
Checks if the password includes spercial character(s)

```python
def check_special(password):
    for special in password:
        if not special.isalnum():
            return True
    return False
```

### Rule 4
Checks if the password includes Uppercase letter(s)

```python
def check_uppercase(password):
    for uppercase in password:
        if uppercase.isupper():
            return True
    return False

```

### General Fuction
Checks the strength of the password.

```python
def check_strength(password):
    score = 0
    if check_length(password):
        score = score + 1
    if check_number(password):
        score = score + 1
    if check_special(password):
        score = score + 1
    if check_uppercase(password):
        score = score + 1
    if score <= 1:
        return "Weak"
    elif score == 2 or score == 3:
        return "Medium"
    else:
        return "Strong"

password = input("Enter your password: ")
print(f"Password strength: {check_strength(password)}")
```
